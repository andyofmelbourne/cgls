```python
def f(x):
    """Here:

    f(x) = 0.5 * xT . A . x - bT . x + c
    A    = [[3, 2], [2, 6]]
    b    = [2, -8]
    c    = 2
    """
    b  = np.array([2.0, -8.0])
    f  = Adot(x)
    f  = 0.5 * np.sum(x * f) - np.sum(b * x) + 2
    return f
    
def Adot(x):
    A = np.array([[3.0, 4.0], [4.0, 6.0]])
    return np.dot(A, x)
```

### Steepest Descent

```python
x0 = np.array([-2.0, -2.0])
sd = cg.Steepest(Adot, np.array([2.0, -8.0]), x0=x0)

xs = []
xs.append(x0)
for i in range(40):
    x = sd.sd(1)
    xs.append(x)
```

```python
N = 100
X, Y = np.meshgrid(np.linspace(-4.0, 25.0, N), np.linspace(-20.0, 4.0, N))
Z = []
for x, y in zip(X.flatten(), Y.flatten()):
    Z.append(f(np.array([x,y])))

Z = np.array(Z).reshape(X.shape)

myShow(X, Y, Z, xs = xs, xdot=None, ydot=None, title = None)
```

### cgls

```python
x0 = np.array([-2.0, -2.0])
cgls = cg.Cgls(Adot, np.array([2.0, -8.0]), x0=x0)

xs_cgls = []
xs_cgls.append(x0)
for i in range(2):
    x = cgls.cgls(1)
    xs_cgls.append(x)
```
