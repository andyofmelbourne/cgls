```python
a = 1.0
b = 1.0
c = 30.0
x0 = 5.0
def f(x):
    f = a * x**2 + c * np.exp( -b * (x - x0)**2)
    return f

def fd(x, d):
    out = d * (2 * a * x - 2 * (x - x0) * c * np.exp( -b * (x - x0)**2))
    return out

def dfd(x, d):
    out = 2 * a - 2 * c * np.exp( -b * (x - x0)**2) + 4 * (x - x0)**2 * c**2 * np.exp( -b * (x - x0)**2)
    out = d * out * d
    return out
```
